# EXTENSION

First, I wanted to look at neutral-aligned heroes, since the main project didn't touch on them:

```python
neutchars = marvel_df.loc[marvel_df['Alignment'] == "neutral"]
neutchars.sort_values(by=['Power', 'Total'], axis=0, ascending=False)
neutchars
```

Since we now have dataframes for each alignment, I decided to compare them:

```python
poweravgs = [np.mean(goodchars['Power']),np.mean(badchars['Power']), np.mean(neutchars['Power']) ]
cols = ["Good", "Bad", "Neutral"]
poweravg_df = pd.DataFrame(cols, columns=["Alignment"])
poweravg_df["Average Power"] = poweravgs


plt.bar(cols, poweravgs)
plt.xlabel("Alignment")
plt.ylabel("Average Power Level")
plt.title("Average Power Level per Alignment")
plt.show()
```

I was surprised at how low the average was for Good heroes, and how high it was for Neutral! I thought this was because of the size per group, as we saw in point 5, so I compared group size to the averages I found: First, I set up a dataframe...

```python
sizes = [432, 165, 11]
poweravg_df['Size'] = sizes
poweravg_df
```

Then, I made the comparison:

```python
sns.set_theme(style="ticks")
sns.lmplot(data=poweravg_df, x="Average Power", y="Size",ci=None)

plt.show()
```

We can see that there is, indeed, a strong negative correlation with group size and average power level!

Would this be the case for other stats? I looked at Total, Speed, and Strength next; Starting with Total:

```python
totalavgs = [np.mean(goodchars['Total']),np.mean(badchars['Total']), np.mean(neutchars['Total']) ]
cols = ["Good", "Bad", "Neutral"]
totalavg_df = pd.DataFrame(cols, columns=["Alignment"])
totalavg_df["Average Total"] = totalavgs

plt.bar(cols, totalavgs)
plt.xlabel("Alignment")
plt.ylabel("Average Total")
plt.title("Average Total per Alignment")
plt.show()
```

```python
sizes = [432, 165, 11]
totalavg_df['Size'] = sizes
totalavg_df
```

```python
sns.set_theme(style="ticks")
sns.lmplot(data=totalavg_df, x="Average Total", y="Size",ci=None)

plt.show()
```

Then, Speed:

```python
speedavgs = [np.mean(goodchars['Speed']),np.mean(badchars['Speed']), np.mean(neutchars['Speed']) ]
cols = ["Good", "Bad", "Neutral"]
speedavg_df = pd.DataFrame(cols, columns=["Alignment"])
speedavg_df["Average Speed"] = speedavgs

plt.bar(cols, speedavgs)
plt.xlabel("Alignment")
plt.ylabel("Average Speed")
plt.title("Average Speed per Alignment")
plt.show()
```

```python
sizes = [432, 165, 11]
speedavg_df['Size'] = sizes
speedavg_df
```

```python
sns.set_theme(style="ticks")
sns.lmplot(data=speedavg_df, x="Average Speed", y="Size",ci=None)

plt.show()
```

And finally, Strength:

```python
strengthavgs = [np.mean(goodchars['Strength']),np.mean(badchars['Strength']), np.mean(neutchars['Strength']) ]
cols = ["Good", "Bad", "Neutral"]
strengthavg_df = pd.DataFrame(cols, columns=["Alignment"])
strengthavg_df["Average Strength"] = strengthavgs

plt.bar(cols, strengthavgs)
plt.xlabel("Alignment")
plt.ylabel("Average Strength")
plt.title("Average Strength per Alignment")
plt.show()
```

```python
sizes = [432, 165, 11]
strengthavg_df['Size'] = sizes
strengthavg_df
```

```python
sns.set_theme(style="ticks")
sns.lmplot(data=strengthavg_df, x="Average Strength", y="Size",ci=None)

plt.show()
```

This appears to be the case for all 4 statistics!

Once again, we can see Neutral-aligned heroes take the higher end of the statistic, since they encompass a smaller range of data. Finally, I decided to look at exactly how small their range is in comparison to Good and Bad:

```python
sns.violinplot(data=goodchars, palette="light:g", inner="points", orient="h")
sns.violinplot(data=neutchars, palette="light:r", inner="points", orient="h")
plt.show()
```

```python
sns.violinplot(data=badchars, palette="light:b", inner="points", orient="h")
sns.violinplot(data=neutchars, palette="light:r", inner="points", orient="h")
plt.show()
```

Here, we can see from the change in the way the plots taper, that the concentration of data is smaller for the Neutral group than Good or Bad, based on how much Green (Good) or Blue (Bad) shows from underneath the Red (Neutral)
